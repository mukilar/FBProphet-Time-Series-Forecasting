"Prophet is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects. It works best with time series that have strong seasonal effects and several seasons of historical data. Prophet is robust to missing data and shifts in the trend, and typically handles outliers well."

## DATA PREPARATION

```python
df1.head()
```

```python
df1_prophet = df1[['Date', 'AveragePrice']]
```

```python
df1_prophet
```

We only need the Date and AveragePrice columns for this time series forecasting. The columns also need to be renamed to fit the Prophet algorithm.

```python
df1_prophet = df1_prophet.rename(columns = {'Date': 'ds' , 'AveragePrice': 'y'})
```

```python
df1_prophet
```

## MODEL IMPLEMENTATION AND FORECASTING

```python
# Fit dataset to the model
model = Prophet()
model.fit(df1_prophet)
```

```python
# Forcasting into the future for the next 365days
future = model.make_future_dataframe(periods = 365)
forecast = model.predict(future)
```

```python
forecast
```

```python
figure = model.plot(forecast, xlabel = 'Date', ylabel = 'Price')
```

```python
figure2 = model.plot_components(forecast)
```

The forecast and trend plot shows that the average price for avocado will follow the same seasonality as the past years in the next 365 days but the overall trend will continue to decrease.

## REGION SPECIFIC FORECASTING

The model predicted the overall trend for avocado price to decrease in the next year. But is this true for all the regions? To check if there is a difference between the overall trend and individual regions, let us use the same model on one particular region.

### West Region

```python
# Select specific region
```

```python
df_sample = df[df['region'] == 'West']
```

```python
df_sample = df_sample.sort_values('Date')
```

```python
plt.plot(df_sample['Date'], df_sample['AveragePrice'])
```

```python
df_sample = df_sample.rename(columns = {'Date':'ds', 'AveragePrice':'y'})
```

```python
m = Prophet()
m.fit(df_sample)
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure3 = m.plot_components(forecast)
```

### Chicago Region

```python
df_sample2 = df[df['region'] == 'Chicago']
```

```python
df_sample2 = df_sample2.sort_values('Date')
```

```python
plt.plot(df_sample2['Date'], df_sample2['AveragePrice'])
```

```python
df_sample2 = df_sample2.rename(columns = {'Date':'ds', 'AveragePrice':'y'})
```

```python
m = Prophet()
m.fit(df_sample2)
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure4 = m.plot_components(forecast)
```

The above two section results show us that even though the overall price for avocados was forecasted to go down, the individual regions West and Chicago have a steady upwards trend in the next 365 days. This shows that the data needs to be divided into smaller groups and the model needs to be implemented seperately on each group for more accurate forecasts.
